```python
# Import pandas
# Create a series of fruits(string) of length 5 and view it
import pandas as pd
d=pd.Series(['Apple','Mango','Strawberry','Orange','Grapes'])
d
```

```python
# Create a series of prices(int) of length 5 and view it
s=pd.Series([100,120,150,50,70])
s
```

```python
# Create a fruit_prices dataframe out of the 2 series you just made and view it
fruit_prices={
    'Fruit':['Apple','Mango','Strawberry','Orange','Grapes'],
    'Price':[100,120,150,50,70]
}
df=pd.DataFrame(fruit_prices)
df
```

```python
import numpy as np
# Find the mean, sum, standard deviation, variance of the prices
d=np.array(df['Price'])
print('mean:',d.mean());print('Sum:',d.sum())
print('std deviation:',d.std());print('Variance:',d.std()**2)
```
